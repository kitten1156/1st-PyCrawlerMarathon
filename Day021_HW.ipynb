## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from bs4 import BeautifulSoup
chrome_options = Options()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-gpu')
browser = webdriver.Chrome(executable_path='./chromedriver',options=chrome_options)
```

```python
from datetime import datetime 
a = "2020/01/18 17:44"
datetime.strptime(a,'%Y/%m/%d %H:%M') 
```

## 1. 取出今天所有的新聞

```python
browser.get("https://www.ettoday.net/news/news-list.htm")
import time
from datetime import datetime 
for i in range(1000):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000*%i);"%i)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    if datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M') > datetime(2020,1,18):
        print(datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M'))
        continue
    else:
        break
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
browser.get("https://www.ettoday.net/news/news-list.htm")
import time
from datetime import datetime 
for i in range(1000):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000*%i);"%i)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    if datetime.now().hour+8 - datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M').hour < 2:
        print(datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M'))
        continue
    else:
        break
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.now().hour+8 - datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M').hour < 2:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser.get("https://www.ettoday.net/news/news-list.htm")
from selenium.webdriver.support.ui import Select
Y = Select(browser.find_element_by_id('selY'))
Y.select_by_value('2020')
M = Select(browser.find_element_by_id('selM'))
M.select_by_value('1')
D = Select(browser.find_element_by_id('selD'))
D.select_by_value('15')
browser.find_element_by_id('button').click()
import time
from datetime import datetime 
for i in range(1000):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000*%i);"%i)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    ttime = datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M').hour
    if  ttime > 14:
        print(datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M'))
        continue
    else:
        break
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    ttime = datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M').hour
    if ttime > 14 and ttime < 18:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
